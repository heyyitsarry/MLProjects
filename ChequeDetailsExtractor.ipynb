```python
def print_content_details(d):
    """
    Recursively search through the dictionary and print only the 'content' values under the 'value' key.
    """
    if isinstance(d, dict):
        for key, value in d.items():
            if key == 'value':
                if isinstance(value, dict):
                    content = value.get('content')
                    if content is not None:
                        print("Content:", content)
            # Recursively search in nested dictionaries or lists
            if isinstance(value, (dict, list)):
                print_content_details(value)
    elif isinstance(d, list):
        for item in d:
            print_content_details(item)
```

```python
# Define variables to hold content values
Name = None
Amt_Words = None
IFSC = None
Date = None
Amt_Number = None
Account_Number = None

# Function to recursively search and store content values
def store_content_values(d):
    global Name, Amt_Words, IFSC, Date, Amt_Number, Account_Number
    if isinstance(d, dict):
        for key, value in d.items():
            if key == 'value':
                if isinstance(value, dict):
                    content = value.get('content', None)
                    if content is not None:
                        if Name is None:
                            Name = content
                        elif Amt_Words is None:
                            Amt_Words = content
                        elif IFSC is None:
                            IFSC = content
                        elif Date is None:
                            Date = content
                        elif Amt_Number is None:
                            Amt_Number = content
                        elif Account_Number is None:
                            Account_Number = content
            # Recursively search in nested dictionaries or lists
            if isinstance(value, (dict, list)):
                store_content_values(value)
    elif isinstance(d, list):
        for item in d:
            store_content_values(item)
```

```python
# Run the function to store content values
store_content_values(result_dict)

# Print the stored content values
print("Name:", Name)
print("Amt_Words:", Amt_Words)
print("IFSC:", IFSC)
print("Date:", Date)
print("Amt_Number:", Amt_Number)
print("Account_Number:", Account_Number)
```
